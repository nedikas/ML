# Anscombe

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset("anscombe")
df.head()
```

```python
df.info()
```

- Le dataset contient 3 colonnes, contenant apparemment 3 variables continues (x, y, dataset)

```python
sns.set_theme(style='darkgrid')
ax = sns.scatterplot(data=df, x="x", y="y");
ax.set_title("Anscombe Dataset. X,Y scatterplot")
plt.show()
```

- L'affichage ne produit pas d'information solides. Les points semblent emmel√©s.

```python
#1. Analysez la relation entre x et y.
ax = sns.scatterplot(data=df, x="x", y="y", hue="dataset")
ax.set_title("Anscombe x,y scatterplot per dataset data");
```

- Pas plus de d√©tails...

```python
#Analyse de la variable dataset
df.dataset.describe()
```

- Seulement 4 valeurs => affichons les

```python
df.dataset.value_counts()
```

- Il s'agit de chiffres romains !! Affichons les donn√©es relativement √† chacune des valeurs de dataset

```python
sns.relplot(data=df, x='x',y='y', col='dataset');
plt.suptitle("Affichages de x,y en fonction des valeurs de dataset");
plt.tight_layout()
plt.show()
```

- Analyse des statistiques de chaque paquet:

```python
df.groupby('dataset').describe()
```

Les quatres paquets contiennet le m√™me nombre de points, ont les m√™mes valeurs de statistiques principales (moyenne, √©cart-type)
